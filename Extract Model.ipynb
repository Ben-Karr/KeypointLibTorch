```python
import torch
import torchvision
```

```python
model = torchvision.models.detection.keypointrcnn_resnet50_fpn(pretrained = True)
model.eval();

model
```

```python
IMG_SIZE = (3, 5:2, 512)
dummy = torch.randn(1, *IMG_SIZE)

traced_net = torch.jit.script(model, dummy)

traced_net.save('models/traced_keypoint_model.pt')
```
